### Only compare the summer seasons
* This is because not all countries compete in the winter season and the medals obtained are always much less. This obscures the trend, so only the olympics in the summer season are considered in the analysis.

```python
s = data.loc[data['Season'] == 'Summer']
```

```python
def get_gdp(country):
    d = gdp[country].to_frame()
    d.columns = ["GDP"]
    d['Country'] = country
    
    return d
```

```python
def get_region(region):
    d = s.loc[data['region'] == region]
    d = d.fillna(d.mean())
    d['total'] = d['Gold'] + d['Silver'] + d['Bronze']
    d = d.groupby(['Year'], as_index=False)[['Gold', 'Silver', 'Bronze', 'total']].sum()
    d['region'] = region

    return d

sports = sorted(data['Sport'].unique().tolist())
```

```python
# unique names of regions    
common_countries = sorted(list(ol_regions.intersection(gdp_countries)))

def modify_doc(doc):
    
    a = "Netherlands"
    b = "Belgium"
    
    source1 = ColumnDataSource(data=get_region(a))
    source2 = ColumnDataSource(data=get_region(b))
    
    source3 = ColumnDataSource(data=get_gdp(a))
    source4 = ColumnDataSource(data=get_gdp(b))
    
    hover_p = HoverTool(
        tooltips=[
            ('Year', "@Year"),
            ('Total', "@total"),
            ('Golds', "@Gold"),
            ('Silvers', "@Silver"),
            ('Bronzes', "@Bronze"),
            ('Country', "@region")
        ]
    )
    
    hover_g = HoverTool(
        tooltips=[
            ('Year', "@Year"),
            ('GDP', "@GDP"),
            ('Country', "@Country")
        ])
    
    p = figure(plot_height=400, 
               plot_width=450,
               tools = [hover_p],
               title = f"Comparing medals for {a} and {b}")
    
    g = figure(plot_height=400,
               plot_width=450,
               tools=[hover_g],
               title=f"Comparing GDP for {a} and {b}")
    
    # title
    p.title.align = "center"
    
    g.title.align = "center"
    
    # x axis
    p.xaxis.axis_label = "Years"
    p.xaxis.axis_label_text_font_style = "bold"
    
    g.xaxis.axis_label = "Years"
    g.xaxis.axis_label_text_font_style = "bold"
    
    # y axis
    p.yaxis.axis_label = "N of medals"
    p.yaxis.axis_label_text_font_style = "bold"
    
    g.yaxis.axis_label = "GDP per capita in US dollars (adjusted)" #divided by (midyear) population
    g.yaxis.axis_label_text_font_style = "bold"

    # Glyph
    p.line('Year', 'total', line_color="#ff8888", line_width=2, alpha=0.7, legend_label="Country 1", source=source1)
    p.line('Year', 'total', line_color='black', line_width=2, alpha=0.7, legend_label="Country 2", source=source2)
    
    g.line('Year', "GDP", line_color="#ff8888", line_width=2, alpha=0.7, legend_label="Country 1", source=source3)
    g.line('Year', "GDP", line_color="black", line_width=2, alpha=0.7, legend_label="Country 2", source=source4)
    
    # Legend
    p.legend.location = "top_left"
    g.legend.location = "top_left"
    
    # Select
    select_c1 = Select(title="Compare country 1:", value="Netherlands", options=common_countries, width=300)
    select_c2 = Select(title="to country 2:", value="Belgium", options=common_countries, width=300)
    
    # Callbacks
    def update_c1(attrname, old, new):
        a = select_c1.value
        b = select_c2.value
        source1.data = get_region(a)
        source3.data = get_gdp(a)
        
        p.title.text = f"Comparing medals for {a} and {b}"
        g.title.text = f"Comparing GDP for {a} and {b}"
    
    select_c1.on_change('value', update_c1)
        
    def update_c2(attrname, old, new):
        a = select_c1.value
        b = select_c2.value
        source2.data = get_region(b)
        source4.data = get_gdp(b)
        
        p.title.text = f"Comparing medals for {a} and {b}"
        g.title.text = f"Comparing GDP for {a} and {b}"
        
    select_c2.on_change('value', update_c2)
        
    # create a layout
    #layout = row(column(select_c1, select_c2, p), row(g))
    layout = column(row(select_c1, select_c2), row(p, g))
    doc.add_root(layout)

show(modify_doc)
```

## Relationship between N of medals and GDP
* Group by country
* From the olympics df, only select the years for which there are measures for GDP
* For every country, sum the n of medals and the gdp (for all the years)
* Plot each country 

```python
# medals in the summer olympics after 1975
meds = s.loc[s['Year'] > 1975]

# GDP for countries with metrics from at least 1975
gd = gdp.loc['1975':'2016'].transpose().dropna()

gd.head(2)
```

```python
gr_meds = meds.groupby(["region"], as_index=False)[['Gold', 'Silver', 'Bronze']].sum()
gr_meds['total'] = gr_meds['Gold'] + gr_meds['Silver'] + gr_meds['Bronze']
gr_meds.rename(columns={'region':'country'}, inplace=True)

gd['sum_gdp'] = gd.sum(axis=1)
gd_n = gd['sum_gdp'].to_frame().reset_index()

in_gdp = set(gd_n.country.unique().tolist())
in_meds = set(gr_meds.country.unique().tolist())

to_drop = list(in_meds.difference(in_gdp))
to_drop2 = list(in_gdp.difference(in_meds))

to_drop.extend(to_drop2)

for i in to_drop:
    gr_meds = gr_meds.drop(gr_meds.index[gr_meds['country'] == i])

gr_meds.head()
```

```python
scatter = pd.merge(gr_meds, gd_n, on='country')
scatter['st_meds'] = scatter['total']

scatter.head(2)
```

```python
def get_sport(sport):
    d = data.loc[data['Sport'] == sport]
    grouped = d.groupby(['Year'])['Age'].mean()
    grouped = grouped.reset_index()
    return grouped

sports = sorted(data['Sport'].unique().tolist())
d = get_sport('Polo')
d
```

```python
def modify_doc(doc):
    
    d = get_sport('Basketball')
    source = ColumnDataSource(data=d)
    
    years = sorted([str(a) for a in d['Year'].unique().tolist()])
    age = d['Age']
    
    hover = HoverTool(
        tooltips=[('Year', "@Year")]
    )

    p = figure(plot_height=400, plot_width=700, title = "Age over the years for Basketball", tools=[hover])
    
    # title
    p.title.text_font_size = '25px'
    p.title.align = "center"
    
    # x axis
    p.xaxis.axis_label = ""
    p.xaxis.axis_label_text_font_style = "bold"
    
    # y axis
    p.yaxis.axis_label = ""
    p.yaxis.axis_label_text_font_style = "bold"

    # scatter with categorical mapping for the regions
    
    p.vbar(x='Year', top='Age', width=0.5, source=source)
    #p.xaxis.major_label_orientation = 1
    
    select = Select(title="Select sport:", value="Basketball", options=sports)
    
    def update_data(attrname, old, new):
        sport = select.value
        source.data = get_sport(sport)
        
        p.add_tools(hover)
        
        p.title.text = f'Age over the years for {sport}'
               
    # link the callback function to the slider's value
    select.on_change('value', update_data)

    # create a layout
    layout = column(row(select), row(p))
    doc.add_root(layout)
    
show(modify_doc)
```
